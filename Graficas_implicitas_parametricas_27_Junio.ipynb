```python
from sympy import sin, symbols, Abs, ln
from sympy.plotting import plot
def f(x):
  return (x + 3 )*(x - 2)*(x + 4)

def g(x):
  return ln(x)

x = symbols('x')

plot(f(x))   # grafica polinomio
plot(f(x), (x, -5, 3), line_color='red')    #cambio el rango
plot(g(f(x)))    #composicion con ln y el polinomio
```

FUNCION IMPLICITA** CON UNA IGUALDAD**

```python
from sympy import symbols, Eq
from sympy import plot_implicit
x, y = symbols('x y')
plot_implicit(Eq(x**4 - (x**2 - y**3), 0), (x, -1.5, 1.5),
              (y, -1.5, 1.5));
```

Funciones Param√©tricas: EXPRESAMOS A X y Y COMO UN PARAMETRO

```python
from sympy import symbols, sin, cos, pi, exp
from sympy import plot_parametric
t = symbols('t')
def x(t):
    return sin(t)*(exp(cos(t)) - 2*cos(4 * t) - sin(t/12)**5)
def y(t):
   return cos(t) * (exp(cos(t)) - 2*cos(4*t) - sin(t/12)**5)

plot_parametric(x(t), y(t), (t, -6*pi, 6*pi));
plot_parametric(x(t), y(t), (t, -50, 50));
plot_parametric(x(t), y(t), (t, -90, 90));
```

Susesi칩n que converge a  ** e**

```python

#sucesi칩n que converge a  e
import numpy as np
e = []

for n in range(1,1000):     #acambiar os valores para 침la aproximacion mas cercana
    aprox =(1 + 1/n)**n
    e.append(aprox)
print(f'{n:7}{aprox:>18.15f}')

print('el valor de e con NUMPY es',np.e)
```

** calcular PI Leibniz**

```python
import numpy as np

pi = []
aprox = 0

for n in range(1,1000):     #acambiar os valores para 침la aproximacion mas cercana
        aprox =aprox + ((-1)**(n+1)*1/(2*n-1))
        pi.append(aprox)
print(f'{n:7}{4*aprox:>18.15f}')

print('el valor de e con NUMPY es',np.e)
```

SERIE DE TAYLOR

```python
from sympy import symbols, cos
x = symbols('x')
cosx = cos(x)
print(cosx.series(x, n= 20))
print(cosx.series(x, n= 20).removeO())
```

```python
from sympy import symbols, ln
x = symbols('x')
lnx = ln(x)
print(lnx.series(x, n= 20,x0=1))
print(lnx.series(x, n= 20, x0=1).removeO())
```
